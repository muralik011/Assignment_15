<b>7. Write a generator, genPrimes, that returns the sequence of prime numbers on successive calls to
its next() method: 2, 3, 5, 7, 11, ...

```python

def prime(n):
    prime_list = []
    for i in range(2, n+1):
        fl = 0
        for j in range(2, i - 1):
            if i % j == 0:
                fl = 1
                break
            
        if fl == 0: 
            prime_list.append(i)
            
    return prime_list
    
n = 1   
def genPrimes():
    global n
    n += 1
    while True:
        if n in prime(n):
            yield(prime(n))    
        else:
            n += 1
            prime(n)
```

```python
next(genPrimes())
```
